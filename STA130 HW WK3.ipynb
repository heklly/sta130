````python
#Q1

import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")
df_adelie = df[df["species"] == "Adelie"]
df_chinstrap = df[df["species"] == "Chinstrap"]
df_gentoo = df[df["species"] == "Gentoo"]

# Calculate statistics
def get_statistics(df, column):
    mean = df[column].mean()
    median = df[column].median()
    std = df[column].std()
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    return mean, median, std, q1, q3

# For Adelie
mean_adelie, median_adelie, std_adelie, q1_adelie, q3_adelie = get_statistics(df_adelie, 'flipper_length_mm')

# For Chinstrap
mean_chinstrap, median_chinstrap, std_chinstrap, q1_chinstrap, q3_chinstrap = get_statistics(df_chinstrap, 'flipper_length_mm')

# For Gentoo
mean_gentoo, median_gentoo, std_gentoo, q1_gentoo, q3_gentoo = get_statistics(df_gentoo, 'flipper_length_mm')

# Create subplots
fig = make_subplots(rows=1, cols=3, subplot_titles=("Adelie", "Chinstrap", "Gentoo"))

# Adelie Histogram
fig.add_trace(go.Histogram(x=df_adelie['flipper_length_mm'], name='Adelie', nbinsx=20, marker_color='lightblue'), row=1, col=1)
# Add vlines for mean, median, and 2 std with adjusted positions
fig.add_vline(x=mean_adelie, line=dict(color='blue', width=2, dash='dash'), annotation_text="Mean", annotation_position="top left", row=1, col=1)
fig.add_vline(x=median_adelie, line=dict(color='green', width=2, dash='dash'), annotation_text="Median", annotation_position="bottom left", row=1, col=1)
fig.add_vline(x=mean_adelie + 2 * std_adelie, line=dict(color='orange', width=2, dash='dash'), annotation_text="+2 Std", annotation_position="top right", row=1, col=1)
fig.add_vline(x=mean_adelie - 2 * std_adelie, line=dict(color='orange', width=2, dash='dash'), annotation_text="-2 Std", annotation_position="bottom left", row=1, col=1)
# Add vrect for IQR
fig.add_vrect(x0=q1_adelie, x1=q3_adelie, fillcolor="purple", opacity=0.3, line_width=0, row=1, col=1, annotation_text="IQR")

# Chinstrap Histogram
fig.add_trace(go.Histogram(x=df_chinstrap['flipper_length_mm'], name='Chinstrap', nbinsx=20, marker_color='lightblue'), row=1, col=2)
# Add vlines for mean, median, and 2 std with adjusted positions
fig.add_vline(x=mean_chinstrap, line=dict(color='blue', width=2, dash='dash'), annotation_text="Mean", annotation_position="top left", row=1, col=2)
fig.add_vline(x=median_chinstrap, line=dict(color='green', width=2, dash='dash'), annotation_text="Median", annotation_position="bottom left", row=1, col=2)
fig.add_vline(x=mean_chinstrap + 2 * std_chinstrap, line=dict(color='orange', width=2, dash='dash'), annotation_text="+2 Std", annotation_position="top right", row=1, col=2)
fig.add_vline(x=mean_chinstrap - 2 * std_chinstrap, line=dict(color='orange', width=2, dash='dash'), annotation_text="-2 Std", annotation_position="bottom left", row=1, col=2)
# Add vrect for IQR
fig.add_vrect(x0=q1_chinstrap, x1=q3_chinstrap, fillcolor="purple", opacity=0.3, line_width=0, row=1, col=2, annotation_text="IQR")

# Gentoo Histogram
fig.add_trace(go.Histogram(x=df_gentoo['flipper_length_mm'], name='Gentoo', nbinsx=20, marker_color='lightblue'), row=1, col=3)
# Add vlines for mean, median, and 2 std with adjusted positions
fig.add_vline(x=mean_gentoo, line=dict(color='blue', width=2, dash='dash'), annotation_text="Mean", annotation_position="top left", row=1, col=3)
fig.add_vline(x=median_gentoo, line=dict(color='green', width=2, dash='dash'), annotation_text="Median", annotation_position="bottom left", row=1, col=3)
fig.add_vline(x=mean_gentoo + 2 * std_gentoo, line=dict(color='orange', width=2, dash='dash'), annotation_text="+2 Std", annotation_position="top right", row=1, col=3)
fig.add_vline(x=mean_gentoo - 2 * std_gentoo, line=dict(color='orange', width=2, dash='dash'), annotation_text="-2 Std", annotation_position="bottom left", row=1, col=3)
# Add vrect for IQR
fig.add_vrect(x0=q1_gentoo, x1=q3_gentoo, fillcolor="purple", opacity=0.3, line_width=0, row=1, col=3, annotation_text="IQR")

# Update layout
fig.update_layout(height=600, width=1000, title_text="Penguin Flipper Length Statistics (Histograms with Summary Statistics) for Adelie, Chinstrap, and Gentoo")

# Show the plot
fig.show(renderer="png")
````

````python
import pandas as pd
import plotly.graph_objects as go
import numpy as np
from scipy import stats

# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")
df = df.dropna(subset=['flipper_length_mm'])  # Drop rows where flipper_length_mm is NaN

# Define a function to calculate statistics
def get_statistics(df, column):
    mean = df[column].mean()
    median = df[column].median()
    std = df[column].std()
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    return mean, median, std, q1, q3

# Create a list of species
species_list = ['Adelie', 'Chinstrap', 'Gentoo']
species_data = [df[df["species"] == species] for species in species_list]

# Loop through each species and create separate plots
for i, species_df in enumerate(species_data):
    mean, median, std, q1, q3 = get_statistics(species_df, 'flipper_length_mm')

    # KDE calculation using scipy.stats.gaussian_kde
    kde = stats.gaussian_kde(species_df['flipper_length_mm'])
    x_vals = np.linspace(species_df['flipper_length_mm'].min(), species_df['flipper_length_mm'].max(), 500)
    y_vals = kde(x_vals)

    # Create a Plotly figure for each species
    fig = go.Figure()

    # Add KDE plot
    fig.add_trace(go.Scatter(x=x_vals, y=y_vals, mode='lines', fill='tozeroy', 
                             name=f'{species_list[i]} KDE', line=dict(color='lightblue')))

    # Add vertical lines for mean, median, and 2 std
    fig.add_vline(x=mean, line=dict(color='blue', dash='dash'), 
                  annotation_text=f"Mean ({species_list[i]})", annotation_position="top left")
    fig.add_vline(x=median, line=dict(color='green', dash='dash'), 
                  annotation_text=f"Median ({species_list[i]})", annotation_position="bottom left")
    fig.add_vline(x=mean + 2 * std, line=dict(color='orange', dash='dash'), 
                  annotation_text=f"+2 Std ({species_list[i]})", annotation_position="top right")
    fig.add_vline(x=mean - 2 * std, line=dict(color='orange', dash='dash'), 
                  annotation_text=f"-2 Std ({species_list[i]})", annotation_position="bottom right")

    # Add shaded region for IQR
    fig.add_vrect(x0=q1, x1=q3, fillcolor='purple', opacity=0.3, 
                  line_width=0, annotation_text=f"IQR ({species_list[i]})")

    # Update layout for each plot
    fig.update_layout(title=f"{species_list[i]} Flipper Length KDE with Summary Statistics",
                      xaxis_title="Flipper Length (mm)", yaxis_title="Density",
                      height=600, width=800)

    # Display the plot using a PNG renderer
    fig.show(renderer="png")
````

````python
bn = pd.read_csv('https://raw.githubusercontent.com/hadley/data-baby-names/master/baby-names.csv')
bn['name'] = bn['name']+" "+bn['sex'] # make identical boy and girl names distinct
bn['rank'] = bn.groupby('year')['percent'].rank(ascending=False)
bn = bn.sort_values(['name','year'])

# the next three lines create the increaes or decrease in name prevalence from the last year 
bn['percent change'] = bn['percent'].diff()
new_name = [True]+list(bn.name[:-1].values!=bn.name[1:].values)
bn.loc[new_name,'percentage change'] = bn.loc[new_name,'percent'] 
bn = bn.sort_values('year')
bn = bn[bn.percent>0.001] # restrict to "common" names
fig2 = px.scatter(bn, x="percent change", y="rank", animation_frame="year", animation_group="name",
                 size="percent", color="sex", hover_name="name",size_max=50, range_x=[-0.005,0.005]) # range_y removed

fig2.update_yaxes(autorange='reversed') # this lets us put rank 1 on the top

fig2.show(renderer="png") # USE `fig.show(renderer="png")` FOR ALL GitHub and MarkUs SUBMISSIONS
````
