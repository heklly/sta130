# Q6


For each variable (column) within a dataframe, DataFrame.describe() provides:

Count - The count summary statistics simply describes how many non-null values are within a column, 
i.e. how many data points there are

Mean - This is the arithmetic mean, the sum of all data points divided by the number of data points, i.e. the average.

Std - The standard deviation, which describes the spread of the data, how far apart they are from the mean.
Higher standard deviation means further apart from the mean.

Min - The minimum value in a column.

25% - The first quartile of the data. Only 25% of the values fall within below the first quartile, which can help 
understand the spread of lower values.

50% - The second quartile of the data, also known as the median value. 50% of the data points are below this, and on the
hand 50% of the values are above

75% - The third quartile of the data, of which 75% of the values are below. Can help understand the spread of the upper values

Max - The maximum value of a column.

## Q7


7.1 df.dropna() simply removes all rows with missing values, while del df['col'] deletes columns.
df.dropna() could discard problematic observations with missing values
instead of entirely discarding possibly important variables. For example, if your variable has a sample of students
and only some students are missing their age, it would be much more useful to drop those students instead of the entire 'age' column.

7.2 On the other hand, del df['col'] discards entire columns, which still has its use cases. If a data set has entirely
useless or unnecessary variables, or missing data points are concentrated in one column, it would be pragmatic to remove the column itself.
For example in a data set containing shoe sales in Canada. If trying to determine the average shoe size of a Canadian, it may be
practical to remove the price of the shoe when analyzing the data set.


7.3 Applying del drop['col'] before df.dropna() is important mainly because it reduces data loss. If a column has many missing values
and df.dropna() is applied first, you could lose a substantial amount of data points, even if the column would've been deleted the step after.
