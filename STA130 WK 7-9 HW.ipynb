### Q1:

A Simple Linear Regression consists of an outcome, 𝑌𝑖 based on a predictor, 𝑥𝑖, as a linear model. It consists of three terms:

$$ 
\begin{align}
Y_i = {}& \beta_0 + \beta_1 x_i + \epsilon_i \quad \text{ where } \quad \epsilon_i \sim \mathcal N\left(0, \sigma^2\right)\\
\end{align}
$$

𝑌𝑖 is the outcome value of the i-th observation. 𝛽0 and 𝛽1 are constant parameters, where 𝛽0 is simply the intercept of the line, and 𝛽1 the slope coefficient.  𝑥𝑖 is the value of the i-th observation. 
𝜖𝑖 is the error term, to account for the inherent noise in the outcome that is not caused by the predictor variable. This is drawn from a normal distribution with parameters variance = 𝜎^2 and mean = 0. The variance in this case is an unobservable constant parameter.

Given this model, this 𝑌𝑖 will follow a normal distrubution due to assumptions about the error term. 𝑌𝑖's distribution can be thought of a deterministic aspect, where the mean is 𝛽0 + 𝛽1𝑥𝑖. The random, normal-distribution aspect of 𝑌𝑖 arises from assuming the error term follows a normal distrubution: N(0,𝜎^2). Since 𝛽0 + 𝛽1𝑥𝑖 is constant, then adding a constant to the normal distrubution of N(0,𝜎^2) results in the following predicted outcome:

$$
\begin{align}
Y_i \sim {}& \mathcal N\left( \beta_0 + \beta_1 x_i, \sigma^2\right)
\end{align}
$$

which itself, is a normal distribution.

````python
import numpy as np
import scipy.stats as stats
import plotly.graph_objects as go

# Parameters for the linear regression model
beta_0 = 2  # Intercept
beta_1 = 1    # Slope
sigma = 0.75     # Standard deviation of the error term

# Updated parameters for the uniform distribution
loc = 0     # Lower bound for x_i
scale = 10  # Width of the uniform distribution

# Generating 100 random points for x_i
np.random.seed(1)
x_i = stats.uniform.rvs(loc=loc, scale=scale, size=100)

# Generating random error terms
epsilon_i = np.random.normal(0, sigma, 100)

# Calculating y_i values
y_i = beta_0 + beta_1 * x_i + epsilon_i

# Creating the plot with Plotly
fig = go.Figure()

# Scatter plot for generated data points
fig.add_trace(go.Scatter(
    x=x_i, y=y_i, mode='markers', name='Generated data points', marker=dict(color='blue')
))

# Line plot for the true regression line (without noise)
fig.add_trace(go.Scatter(
    x=x_i, y=beta_0 + beta_1 * x_i, mode='lines', name='True regression line', line=dict(color='red')
))

# Updating layout
fig.update_layout(
    title="Generated Data Points and True Regression Line",
    xaxis_title="x",
    yaxis_title="y",
    height=600, width=1000,
    xaxis=dict(title="x", range=[0, max(x_i)]),  # Setting x-axis lower bound to 0
    yaxis=dict(title="y", range=[0, max(y_i) + 0.5]),
    legend=dict(title="Legend")
)

fig.show(renderer="png")
````

The differences in the two lines reflect the randomness of sampling and of the error term in Simple Linear Regression model. The true regression line represents an ideal, where the population outcome is entirely determined by the predictor as a proportional relationship. The linear fit model tries to fit a possible linear trendline to a sample, based on OLS. However, as our sample is based on the theoretical model, there is an extra error term added onto our outcome. 

Because of the nature of the error term, assumed to follow a normal distribution, it's subject to inherent randomness, and so our sample data set simulated under the Simple Linear Regression model will vary from one sample data set to another, even for the exact same xi value. Thus, the linear fit for the simulated data set will vary from one another, and this is reflected in the fit not exactly matching the true regression line -- similar to how a sample might not be a true, exact representation of the population. 

In addition, since the linear fit model is based on OLS, which tries to find the best-fitting line by minimizing the sum of squared residuals (the squared difference between observed and predicted values), if the datapoints includes outliers, it can skew the line.

### Ethical Professionalism Considerations
    
The TUT and HW both addressed some of the assumptions used in **Simple Linear Regression**. The **p-values** provided by `statsmodels` via `smf.ols(...).fit()` depend on these assumptions, so if they are not (at least approximately) correct, the **p-values** (and any subsequent claims regarding the "evidience against" the **null hypothesis**) are not reliable. In light of this consideration, describe how you could diagnostically check the first three assumptions (given below) when using analyses based on **Simple Linear regression** model. From an Ethical and Professional perspective, do you think doing diagnostic checks on the assumptions of a **Simple Linear regression** model is something you can and should do whenever you're doing this kind of analysis? 
            
> The first three assumptions associated with the **Simple Linear regression** model are that
> 
> - the $\epsilon_i$ **errors** (sometimes referred to as the **noise**) are **normally distributed**
> - the $\epsilon_i$ **errors** are **homoscedastic** (so their distributional variance $\sigma^2$ does not change as a function of $x_i$)
> - the linear form is [at least reasonably approximately] "true" (in the sense that the above two remain [at least reasonably approximately] "true") so that then behavior of the $Y_i$ **outcomes** are represented/determined on average by the **linear equation**)<br>
> 
>    and there are additional assumptions; but, a deeper reflection on these is "beyond the scope" of STA130; nonetheless, they are that<br><br>
> - the $x_i$ **predictor variable** is **measured without error**
> - and the $\epsilon_i$ **errors** are **statistically independent** (so their values do not depend on each other)
> - and the $\epsilon_i$ **errors** are **unbiased** relative to the **expected value** of **outcome** $E[Y_i|x_i]=\beta_0 + \beta_1x_i$ (which is equivalently stated by saying that the mean of the **error distribution** is $0$, or again equivalently, that the **expected value** of the **errors** $E[\epsilon_i] = 0$)
