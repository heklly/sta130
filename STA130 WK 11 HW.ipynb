1. **Accuracy** measures the proportion of true results (both true positives and true negatives) in the population.

$$\text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN}$$
    
2. **Sensitivity** measures the proportion of actual positives that are correctly identified.

$$\text{Sensitivity} = \frac{TP}{TP + FN}$$

3. **Specificity** measures the proportion of actual negatives that are correctly identified.

$$\text{Specificity} = \frac{TN}{TN + FP}$$

4. **Precision** measures the proportion of positive identifications that were actually correct.

$$\text{Precision} = \frac{TP}{TP + FP}$$

Accuracy just describes how many "correct" predictions a model makes divided by the total number of observations. A "correct" prediction is if the model predicts positive and it is positive, or if the model predicts negative and it is negative. For example, when describing the performance of a medical test, accuracy might be approriate. A pregnancy test might claim a 90% accuracy rate. This would mean that when a person is pregnant it predicts they are 90% of the time. If they aren't, it also predicts they aren't pregnant 90% of the time. It is a straightforward metric to get a general sense of how well a model performs.


Sensitivity measures out of all the positives, how many positives were correctly identified? Sensitivity is highly relevant in an airport security context. Missing out on contraband could lead to disastrous consequences on a plane, while a false positive may only lead to annoyance and delay.


Specificity refers to out of all the negatives, how many negatives were correctly identified? For example, In a criminal justice system, high specificity is vital. High specifity in this case would mean someone who is innocent is correctly labelled innocent. Hence, the rule of "beyond a reasonable doubt", as identifying someone who was innocent as guilty has ramifications both for the victim and trust in the legal system.


There is usually a tradeoff between specificity and sensitivity. If a test has high sensitivity, tuned to identify positives correctly, it's likely that some negatives will be wrongfully identified as positive and vice versa.


Precision measures the proportion of true positives out of all positive test results. Precision might be more important when you have to be highly confident in your positive results. For example, in a diagnostic test for leukemia, precision must be prioritized because the treatments for leukemia are invasive, and subjecting someone who tested positive but is actually not could lead to a lot of side effects and psychological distress.

##### **Key Concepts Discussed:**

1. **Classification Decision Tree:**
   - **What it addresses:** Classification trees solve problems by categorizing data into predefined classes based on feature values.
   - **Example:** A tree predicting whether a loan application is approved or denied based on applicant features like income and credit score.

2. **Classification Trees vs. Multiple Linear Regression:**
   - **Classification Trees** predict discrete outcomes (classes), splitting data based on decision rules at each node.
   - **Multiple Linear Regression** predicts continuous outcomes by fitting a linear relationship between input variables and the target variable.

3. **Accuracy as a Metric for Classification Trees:**
   - **Definition:** Accuracy measures the proportion of correct predictions (both positive and negative) in a classification task.
   - **When important:** Accuracy is crucial when classes are balanced, and false positives and false negatives are equally problematic.

4. **Sensitivity and Specificity:**
   - **Sensitivity (True Positive Rate):** Measures the proportion of actual positives correctly identified by the model.
     - **When important:** Sensitivity is crucial when the **cost of false negatives** is high (e.g., in medical diagnoses, fraud detection).
   - **Specificity (True Negative Rate):** Measures the proportion of actual negatives correctly identified.
     - **When important:** Specificity matters when the **cost of false positives** is high (e.g., in medical testing, ensuring that healthy people are not misdiagnosed).

5. **Precision:**
   - **Definition:** Precision measures how accurate the model is when it predicts a positive outcome (i.e., how many of the predicted positives are truly positive).
   - **When important:** Precision is key when **false positives** are costly or disruptive (e.g., fraud detection, spam filtering).
   - **Contrasting with Specificity:** Both metrics aim to reduce false positives, but **precision** focuses on the **positive predictions** made, while **specificity** focuses on **actual negatives**.

6. **Concrete Example: Email Spam Detection:**
   - **Precision vs. Specificity:** 
     - **Precision**: Focuses on ensuring that when an email is marked as spam, it’s truly spam (important in avoiding false positives).
     - **Specificity**: Focuses on ensuring that legitimate emails are not wrongly marked as spam (important in minimizing false positives among actual non-spam emails).

### **Summary of Key Differences:**
- **Precision**: Reduces false positives in the **predicted positives** (e.g., spam emails marked as spam).
- **Specificity**: Reduces false positives in the **actual negatives** (e.g., legitimate emails wrongly marked as spam).

<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>
  
> The way a **classification decision tree** is fit is that at each step in the construction process of adding a new **decision node splitting rule** to the current **tree structure**, all possible **decision rules** for all possible **predictor variables** are considered, and the combination that improves the **prediction** the most (as measured by the criterion of either "Gini impurity" or "Shannon entropy") and in accordance with the rules of the decision tree (such as the `max_depth` argument) is added to the **classification decision tree**.  Thus overall "criterion" noted above improves with each new **decision node splitting rule**, so the improvement can thus be tracked and the improvement contributions attributed to the **feature** upon which the **decision node splitting rule** is based.  This means the relative contribution of each **predictor variable** to the overall explanatory power of the model can be calculated, and this is what the `.feature_importances_` attribute does. 
>
> Compared to the simplicity of understanding how different **covariates** contribute towards the final **predicted values** of **multiple linear regression models** (by just reading off the equation to see how predictions work), the the complexity of how all the different **features** interact and combine to together to create the final **predictions** from **classification decision trees** can be staggering. But the so-called **feature importance** heuristics allows us to judge how relatively important the overall contributions from different features are in the final decision tree predictions. Now we just need to be sure we're not **overfitting** our **classification decision trees** since they can be so **complex**. Fortunately, the "GridSearchCV" methodology mentioned in regards to finding the best `max_depth` setting for a tree is going to provide a general answer to the challenge of complexity and **overfitting** in **machine learning models** that is not too hard to understand (and which you might already have some guesses or a hunch about).
