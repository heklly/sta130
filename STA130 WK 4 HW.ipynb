#### Q8 Methodology Code and Explanations:

To calculate a confidence interval, I'll bootstrap my original sample numerous times (in this case 2500 times) and from that new bootstrapped means data set, calculate which values fall within the 95% range of that. That interval will become my confidence intervals for the mean.

I've visualized the bootstrapped means as overlying histograms of each other, that way I can get a visual sense of how much the confidence intervals would overlap directly over the bootstrapped means. After that visualization, I transfer that confidence interval to the original box plots. Printed below that is the exact values for the confidence interval.

````python
import pandas as pd
import plotly.graph_objects as go
import numpy as np

# Load the data
url = "https://raw.githubusercontent.com/heklly/sta130/refs/heads/main/Alitech%20Vaccine%20Data.csv"
data = pd.read_csv(url)

np.random.seed(1001)
# Function to perform bootstrapping and return a dataset of sample means
def bootstrap_sample_means(data_column, n_iterations=1000, sample_size=None):
    """
    Perform bootstrapping to generate a dataset of sample means.
    
    Parameters:
    - data_column: The column of data to bootstrap from (Pandas Series or NumPy array).
    - n_iterations: The number of bootstrap iterations (default is 1000).
    - sample_size: The size of each bootstrap sample (default is the size of the original dataset).
    
    Returns:
    - A list of bootstrapped sample means.
    """
    if sample_size is None:
        sample_size = len(data_column)  # Default to the size of the original dataset

    bootstrapped_means = []
    
    # Perform bootstrapping
    for i in range(n_iterations):
        # Take a random sample (with replacement) from the data
        sample = np.random.choice(data_column, size=sample_size, replace=True)
        # Compute the mean of the sample
        sample_mean = np.mean(sample)
        # Store the mean
        bootstrapped_means.append(sample_mean)
    
    return bootstrapped_means

# Perform bootstrapping on both Initial and Final Health Scores
initial_health_means = bootstrap_sample_means(data['InitialHealthScore'], n_iterations=2500)
final_health_means = bootstrap_sample_means(data['FinalHealthScore'], n_iterations=2500)

# Create a DataFrame of the bootstrapped sample means
bootstrapped_data = pd.DataFrame({
    'InitialHealthScore_Means': initial_health_means,
    'FinalHealthScore_Means': final_health_means
})

#Calculate the confidence interval
initial_health_ci = np.percentile(initial_health_means, [2.5, 97.5])
final_health_ci = np.percentile(final_health_means, [2.5, 97.5])
````

````python
fig = go.Figure()

# Add histogram for Initial Health Score bootstrapped means
fig.add_trace(go.Histogram(
    x=initial_health_means,
    name='Initial Health Score Means',
    marker_color='blue',
    opacity=0.6
))

# Add histogram for Final Health Score bootstrapped means
fig.add_trace(go.Histogram(
    x=final_health_means,
    name='Final Health Score Means',
    marker_color='green',
    opacity=0.6
))

fig.add_vrect(
    x0=initial_health_ci[0], x1=initial_health_ci[1],
    fillcolor="blue", opacity=0.2, line_width=0,
    annotation_text="95% CI (Initial)", annotation_position="top left"
)

fig.add_vrect(
    x0=final_health_ci[0], x1=final_health_ci[1],
    fillcolor="green", opacity=0.2, line_width=0,
    annotation_text="95% CI (Final)", annotation_position="top right"
)

# Update layout for better visualization
fig.update_layout(
    title='Bootstrapped Means for Initial and Final Health Scores',
    xaxis_title='Sample Means',
    yaxis_title='Frequency',
    barmode='overlay',  # Overlay both histograms
    bargap=0,
    height=600, width=1200,
)


fig.show(renderer='png')
````
